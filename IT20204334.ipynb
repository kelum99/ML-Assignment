```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("mushrooms.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['class'].value_counts()
```

```python
data = data.astype('category')
data.dtypes
```

```python
labelencoder=LabelEncoder()
for column in data.columns:
    data[column] = labelencoder.fit_transform(data[column])
```

```python
data.head()
```

```python
data['veil-type']
```

```python
data = data.drop(["veil-type"],axis=1)
data.head()
```

```python
x = data.iloc[:,1:22]
x.head()
```

```python
y = data.iloc[:,0]
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x);
x[0:10]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.svm import SVC
model = SVC(kernel='poly', C=10)
model.fit(x_train, y_train)
```

```python
model.kernel
```

```python
pred = model.predict(x_test)
pred[0:5]
```

```python
y_test[0:5]
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(pred, y_test)
```

```python
confusion_matrix(pred, y_test)
```

```python
from sklearn.svm import SVC

best_Cvalue = 0
best_score = 0
for C in [0.1, 1, 10, 100]:
    svm = SVC(C=C)
    svm.fit(x_train, y_train)
    if svm.score(x_test, y_test) > best_score:
        best_score = svm.score(x_test, y_test)
        best_Cvalue = C
        
print("Best SVM C value: {}".format(best_Cvalue))
print("Test Accuracy: {}%".format(round(best_score*100,2)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix 
y_pred_svm = svm.predict(x_test)
print("SVM Classifier report: \n\n", classification_report(y_test, y_pred_svm))
```

```python
import matplotlib.pyplot as splot
import seaborn

cmatrix = confusion_matrix(y_test, y_pred_svm)
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]
f, ax = splot.subplots(figsize =(7,7))
seaborn.heatmap(cmatrix, annot = True, linewidths=0.2, linecolor="blue", fmt = ".0f", ax=ax, cmap="Purples", xticklabels=x_axis_labels, yticklabels=y_axis_labels)
splot.xlabel("PREDICTED LABEL")
splot.ylabel("TRUE LABEL")
splot.title('Confusion Matrix for SVM Classifier')
splot.show()
```
